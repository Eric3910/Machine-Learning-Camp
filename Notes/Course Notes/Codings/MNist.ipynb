```python
import tensorflow as tf
import numpy as np
import random
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot = True)

# set up attributes
trainNum = 55000
# Total of train pics
testNum = 10000
# Total of test pics
trainSize = 500
# No. of train pics 
testSize = 5
# No. of test pics
k = 4
# k 

trainIndex = np.random.choice(trainNum, trainSize, replace = False)
# in trainNum dataSet, generate trainSize none-dup random nubmers 
testIndex = np.random.choice(testNum, testSize, replace = False)
# in testNum dataSet, generate testSize none-dup random nubmers 
trainData = mnist.train.images[trainIndex]
# pics in training dataset is mnist.train.images
trainLabel = mnist.train.labels[trainIndex]
# labels in training dataset
testData = mnist.test.images[testIndex]
# pics in test dataset 
testLabel = mnist.test.labels[testIndex]
# labels in test dataset
#print('trainData.shape=', trainData.shape)
#print('trainlabel.shape=', trainLabel.shape)
#print('testData.shape=',testData.shape)
#print('testLabel.shape=',testLabel.shape)
#print('testLabel=',testLabel)

#tf.placeholder(dtype, shape = None, name = None)
''' 形式参数，用于定义过程，执行时具体赋值
parameters:
dtype: data type. normally tf.float32, tf.float64, etc.
shape： data shape, default is none, which means one-dimension, could be multi-dimension
name: name 

'''
trainDataInput = tf.placeholder(shape = [None,784], dtype = tf.float32)
trainLabelInput = tf.placeholder(shape = [None,10], dtype = tf.float32)
testDataInput = tf.placeholder(shape = [None,784], dtype = tf.float32)
testLabelInput = tf.placeholder(shape = [None,10], dtype = tf.float32)

# 计算曼哈顿距离

f1 = tf.expand_dims(testDataInput, 1)
f2 = tf.subtract(trainDataInput, f1)
f3 = tf.reduce_sum(tf.abs(f2), reduction_indices = 2)
# tf.reduce_sum is used to ++data

with tf.Session() as sess:
    p1 = sess.run(f1, feed_dict = {testDataInput : testData[0:5]})
    print('p1 = ',p1.shape)
    #p1 = (5,1,784)
    p2 = sess.run(f2,feed_dict = {trainDataInput : trainData, testDataInput: testData[0:5]})
    print('p2 = ',p2.shape)
    #p2 = (5,500,784)
    p3 = sess.run(f3,feed_dict = {trainDataInput : trainData, testDataInput: testData[0:5]})
    print('p3 = ',p3.shape)
    #p3 = (5,500)
    print('p3[0,0]=',p3[0,0])
    #p3[0,0] means the distance between the first test picture with the first train pic
    
#tf.nn.top_k(input, k, name = None)
'''parameters:
input: a tensor, data type should be like float32,float 64,int32,int64,uint8,int16,int8. data-dim is batch_size multiple 
    x types.
k: integer which >=1
name: name

Output:
    Tensor tuple: data elements: (values,indices)
        values: a tensor, data type same with input. data-dim is batch_size multiple k
        indices: a tensor, data type int32, the indices of every maximum value in input

'''

f4 = tf.negative(f3)
f5,f6 = tf.nn.top_k(f4, k = 4)
# f5: values, f6: indices
with tf.Session() as sess:
    p4 = sess.run(f4, feed_dict = {trainDataInput:trainData, testDataInput:testData[0:5]})
    print('p4 = ',p4.shape)
    print('p4[0,0]=',p4[0,0])
    p5,p6 = sess.run((f5,f6), feed_dict = {trainDataInput:trainData, testDataInput: testData[0:5]})
    print('p5 = ',p5.shape)
    print('p6 = ',p5.shape)
    print('p5',p5)
    print('p6',p6)

#tf.gather()
'''
collecting slices from parameter axis according to indices.
Indices should be integer tensor with any dimensions, usually 0-D or 1-D.
The output tensor generated should be shaped like: params.shape[:axis] + indices.shape + params.shape[axis+1:]

Input:
params: tensor,used to collect values.Rank of this tensor should be at least axis +1
indices: tensor, should be type: int32, int64. indices should be in the range of [0, params.shape[axis]]
axis: tensor. int32 or int64. default is the first dimension. support minus indices.
name: name(optional)

Output:
tensor. same type as params. the paras come from the indices. shape: params.shape[:axis] + indices.shape + params.shape[axis+1:]
'''
f7 = tf.gather(trainLabelInput,f6)
# find label value according to the indices.
f8 = tf.reduce_sum(f7,reduction_indices=1)
# sum values of dimension 1
f9 = tf.argmax(f8,dimension=1)
# return the maximum value`s indice
with tf.Session() as sess:
    p7 = sess.run(f7,feed_dict={trainDataInput:trainData,testDataInput:testData[0:5],trainLabelInput:trainLabel})
    print('p7=',p7.shape)
    print('p7[]',p7)
    p8 = sess.run(f8,feed_dict={trainDataInput:trainData,testDataInput:testData[0:5],trainLabelInput:trainLabel})
    print('p8=',p8.shape)
    print('p8[]=',p8)
    p9 = sess.run(f9,feed_dict={trainDataInput:trainData,testDataInput:testData[0:5],trainLabelInput:trainLabel})
    print('p9=',p9.shape)
    print('p9[]=',p9)
    
    
with tf.Session() as sess:
    p10 = np.argmax(testLabel[0:5],axis=1)
 # p9=p10，代表正确
    print('p10[]=',p10)
j = 0
for i in range(0,5):
    if p10[i] == p9[i]:
        j = j+1
print('ac=',j*100/5)
# 正确率

```
