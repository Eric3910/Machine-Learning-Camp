### 循环神经网络
- 语音识别
- 自然语言处理，词语分离
- 小爱，关键字跳起。语音转文字后，通过正则表达式找到对应指令。无上下文语境。
- 循环神经网络在语音或文字输入后，尝试循环理解，一个字，两个字，三个字，知道找到合理的拆分方式，符合正向反馈。正向反馈来自于训练模型。
- 人类智能的关键在于人类对于选择的正向刺激不是立竿见影的。机器智能依靠立刻的正向或负向反馈
- 长短记忆体循环神经网络，LSTM，缓存数据和循环迭代长短。
- 神经网络搭通或训练好以后，也不是一条确定的神经链路，而是一个有向图，每一条边的概率不同，收益也不同。正向激励就是调整某一条神经链路的概率
- 性价比计算（个人计算）价值换算成时间。

#### 机器学习算法
##### 监督学习
- 持续给予带标签的数据进行训练
- 对训练过程进行人工干预（识别结果，调参）
##### 非监督学习
- 只提供数据，不提供标签
- 让机器自行得出结论，某些数据特性相近，或得出某种规律。
##### 半监督学习
- 结合
- 少量有标签的数据和大量无标签的数据进行分类
##### 强化学习
- 适用于有穷解
- 结果好度量
##### 遗传算法
- 基于前一代算法进行进化

### 人类神经网络与人工神经网络

#### 机器神经网络暂且无法直接传递过程中的信号，需要将结果按某种形式作为中转


#### 人工神经网络：
- 所有神经元层级和链接关系无法更换。
- 通过调整个别神经的权重影响结果
- 误差传递增益：误差被当做数据进行处理，所以负反馈也被传递
